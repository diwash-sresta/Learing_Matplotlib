Matplotlib Plotting functions

```python
import matplotlib.pyplot as plt

months = [1,2,3,4]
sales = [1000,2000,1500,1800]

plt.plot(months,sales,color='blue', linestyle='--', linewidth='2', marker='o' ,label='2025 sales data')
plt.xlabel("Months")#for labeling in x-axis
plt.ylabel("Sales Per Month")#for labeling in y-axis
plt.title("Monthly Sales Data Report")
plt.legend(loc = 'upper left' ,fontsize = 10)
plt.grid(color = 'gray', linestyle = ':', linewidth = '1')#for making graph
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])#changing x-axis label 
plt.show()
```
