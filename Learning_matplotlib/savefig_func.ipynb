Example:

```python
import matplotlib.pyplot as plt

x= [1,2,3,4]
y = [10,20,15,25]

#create plot
plt.plot(x,y, marker = 'o')
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.savefig('line_plot.png', dpi= 300, bbox_inches = 'tight' )
plt.show()
```
