Different Types of Charts 

#1 Bar Graph

```python
import matplotlib.pyplot as plt

product = ['A','B','C','D']
sales = [1000,2000,800,1200]

plt.bar(product,sales, color='orange',label = 'sales of the product')#plt.barh = for horizontal 
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Sales of the products')
plt.legend()
plt.show()
  
```

#2 Pie Charts

```python
Regions = ['North','South','East','West']
revenue = [3000,4000,1500,2000]

plt.pie(revenue,labels=Regions,autopct='%1.1f%%',colors=['yellow','orange','red','skyblue'])
plt.title('Revenue Contribution by Region')
plt.legend(loc='upper right')
plt.show()
```

#3 Histograms

```python
scores = [45,67,89,56,79,88,60,92,75,85,75,77,68,88,66]

plt.hist(scores,bins=5,color = 'purple', edgecolor='black')
plt.xlabel("Score Range")
plt.ylabel('Number of the Students')
plt.title('Score Distribution of Students')
plt.show()
```
