Scatter Plot

```python
import matplotlib.pyplot as plt

hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,75,80,85]

#scatter plot
plt.scatter(hours_studies,exam_scores,color='green', marker='o', label='Student data')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Relationship betwn Study hours VS Exam Scores')
plt.legend()
plt.grid(True)
plt.show()
```

Comparing two Scatter plot

```python

plt.scatter([1,2,3],[50,55,60],color='blue', marker='o', label='Student A')
plt.scatter([1,2,3],[45,50,55],color='orange', marker='s', label='Student B')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Relationship betwn Study hours VS Exam Scores')
plt.legend()
plt.grid(True)
plt.show()
```
