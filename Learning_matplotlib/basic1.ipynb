```python
x= ['Sun','Mon','Tues','Wed','Thur','Fri']
y= [10,15,7,20,12,18]

plt.plot(x,y)
plt.title('Bakery Sales This Week')
plt.xlabel('Day of the Week')
plt.ylabel('Sales of the week')
plt.show()
```
